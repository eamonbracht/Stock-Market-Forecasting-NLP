```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import threading 
import csv
def scrape(query, startdate, enddate):
    """
        Scrapes webpage given root and returns array of web elements
        
        Parameters
        ----------
        query       : search query as string
        startdate   : date of first tweet as <year>-<month>-<day> ex: 2018-06-01
        enddate     : date of last tweet as <year>-<month>-<day>
        
        Returns
        -------
        Tweets as web elemtns
        """
    browser = webdriver.Chrome()
    base_url = u'https://twitter.com/search?l=&q='
    sub_url = u"%24{}%20OR%20%23%24{} %20since%3A{} %20until%3A{}&src=typd&lang=en".format(query, query, startdate, enddate) 
    url = base_url+sub_url
    
    browser.get(url)
    time.sleep(1)
    body = browser.find_element_by_tag_name('body')
    for _ in range(100):
        body.send_keys(Keys.PAGE_DOWN)
        time.sleep(0.2)
    
    tweets = browser.find_elements_by_class_name('js-tweet-text-container')
    dates = browser.find_elements_by_class_name('time')
    

    return tweets, dates

def readTextFile(filePath):
    """
        reads in and returns text file as list of strings
        
        Parameters
        ----------
        filePath    : directory path
        
        Returns
        -------
        List of strings
        """
    text_file = open(filepath, "r")
    list = text_file.readlines()
    return list

def saveTweets(tweets, dates, fileName, ticker):
    """
        saves scrapped tweets to .txt file
        
        Parameters
        ----------
        tweets      : list of web elements from scrape
        dates       : dates of each tweet
        fileName    : name file will be written to as a string
        
        """
    out_twitter = []
    for i in range(len(tweets)):
            out_twitter.append([dates[i].text, tweets[i].text])
    
            
    with open(fileName+".csv", 'w') as csvFile:
        writer = csv.writer(csvFile)
        writer.writerows(out_twitter)
        csvFile.close()
        print(" saved {} for {} to {} ".format(len(out_twitter), ticker, fileName))


def proto_exec(tweet_args, filepath): 
    twitter, dates = scrape(*tweet_args) 
    saveTweets(twitter, dates, filepath, ticker) 

start_year = 2009
end_year = 2018
# ticker = "MMM"
# ticker = "BAYRY"
# ticker = "HON"
ticker = "SYF"
threads = [threading.Thread(target=proto_exec, 
            args=((ticker, "{}-01-01".format(start_year + i), "{}-12-31".format(start_year + i)), "{}_{}".format(ticker,start_year + i),))
            for i in range(end_year-start_year+1)
          ]

for thread in threads: 
    thread.start() 

for thread in threads: 
    thread.join()
```
